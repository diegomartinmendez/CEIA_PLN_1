```python
# Limpio el .txt por completo, para que me quede solamente el libro:

# Leer archivo completo:
with open("pg5200.txt", encoding="utf-8") as f:
    raw_text = f.read()

# Busco el marcador de inicio (la primera aparición de "I" como capítulo):
start_marker = "\nI\n"  # 'I' rodeado de saltos de línea, para mayor precisión
start_idx = raw_text.find(start_marker)
if start_idx == -1:
    # Si no está rodeado de \n, busco simplemente "I"
    start_idx = raw_text.find("I")
    if start_idx == -1:
        raise ValueError("No se encontró el marcador de inicio 'I'")

start_idx += len("I")  # saltar la 'I' misma

# Busco marcador de fin:
end_marker = "*** END OF THE PROJECT GUTENBERG EBOOK METAMORPHOSIS ***"     # Esta frase la buscqué en el texto para identificarla, y la pegué en el código. Fue la única forma que encontré para limpiar el .txt.
end_idx = raw_text.find(end_marker)
if end_idx == -1:
    raise ValueError("No se encontró el marcador final del libro.")

# Extrae el texto principal:
clean_text = raw_text[start_idx:end_idx].strip()

# Un ejemplo:
print("Texto limpio extraído:")
print(clean_text[:1000])
```

```python
import pandas as pd
import nltk
# Veo que se hayan descargado los datos:
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize

# Suponiendo que 'clean_text' es string:
oraciones = sent_tokenize(clean_text)

# Creo DataFrame con una oración por fila:
df = pd.DataFrame(oraciones, columns=["oracion"])

# Elimino espacios extra si los hay:
df["oracion"] = df["oracion"].str.strip()

# Eliminar filas vacías, por precaución:
df = df[df["oracion"] != ""].reset_index(drop=True)

# Un ejemplo:
df.head()
```

```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence

sentence_tokens = []
# Recorrer todas las filas y transformar las oraciones en una secuencia de palabras (esto podría realizarse con NLTK o spaCy también):
for _, row in df[:None].iterrows():
    sentence_tokens.append(text_to_word_sequence(row[0]))
```

```python
# Veo un par de oraciones:
sentence_tokens[:2]
```

### Crear los vectores (word2vec)

```python
from gensim.models.callbacks import CallbackAny2Vec
# Durante el entrenamiento gensim por defecto no informa el "loss" en cada época.
# Sobrecargamos el callback para poder tener esta información.
class callback(CallbackAny2Vec):
    """
    Callback to print loss after each epoch
    """
    def __init__(self):
        self.epoch = 0

    def on_epoch_end(self, model):
        loss = model.get_latest_training_loss()
        if self.epoch == 0:
            print('Loss after epoch {}: {}'.format(self.epoch, loss))
        else:
            print('Loss after epoch {}: {}'.format(self.epoch, loss- self.loss_previous_step))
        self.epoch += 1
        self.loss_previous_step = loss
```

```python
# Modelo generador de vectores

# Se utiliza la estructura modelo Skipgram:
w2v_model = Word2Vec(min_count=5,         # frecuencia mínima de palabra para incluirla en el vocabulario
                     window=2,            # cant de palabras antes y desp de la predicha
                     vector_size=300,     # dimensionalidad de los vectores
                     negative=20,         # cantidad de negative samples... 0 es no se usa
                     workers=1,           # si tienen más cores pueden cambiar este valor
                     sg=1)                # modelo 0:CBOW  1:skipgram
```

```python
# Obtener el vocabulario con los tokens:
w2v_model.build_vocab(sentence_tokens)
```

```python
# Cantidad de filas/docs encontradas en el corpus:
print("Cantidad de docs en el corpus:", w2v_model.corpus_count)
```

```python
# Cantidad de words encontradas en el corpus:
print("Cantidad de words distintas en el corpus:", len(w2v_model.wv.index_to_key))
```

### Entrenar embeddings

```python
# Entrenamiento  del modelo generador de vectores:

# Utilizamos nuestro callback:
w2v_model.train(sentence_tokens,
                 total_examples=w2v_model.corpus_count,
                 epochs=20,
                 compute_loss = True,
                 callbacks=[callback()]
                 )
```

### Ensayar

```python
# Palabras que MÁS se relacionan con...:
w2v_model.wv.most_similar(positive=["door"], topn=10)
```

```python
# Palabras que MENOS se relacionan con...:
w2v_model.wv.most_similar(negative=["flat"], topn=10)
```

```python
# el método 'get_vector' permite obtener los vectores:
vector_door = w2v_model.wv.get_vector("door")
print(vector_door)
```

```python
# El método 'most_similar' también permite comparar a partir de vectores:
w2v_model.wv.most_similar(vector_door)
```

### Visualizar agrupación de vectores


```python
from sklearn.decomposition import IncrementalPCA
from sklearn.manifold import TSNE
import numpy as np

def reduce_dimensions(model, num_dimensions = 2 ):

    vectors = np.asarray(model.wv.vectors)
    labels = np.asarray(model.wv.index_to_key)

    tsne = TSNE(n_components=num_dimensions, random_state=0)
    vectors = tsne.fit_transform(vectors)

    return vectors, labels
```

```python
# Graficar los embedddings en 2D:

import plotly.graph_objects as go
import plotly.express as px

vecs, labels = reduce_dimensions(w2v_model)

MAX_WORDS=200
fig = px.scatter(x=vecs[:MAX_WORDS,0], y=vecs[:MAX_WORDS,1], text=labels[:MAX_WORDS])
fig.show(renderer="colab")     # esto para plotly en colab
```

```python
# También se pueden guardar los vectores y labels como tsv para graficar en
# http://projector.tensorflow.org/

vectors = np.asarray(w2v_model.wv.vectors)
labels = list(w2v_model.wv.index_to_key)

np.savetxt("vectors.tsv", vectors, delimiter="\t")

with open("labels.tsv", "w") as fp:
    for item in labels:
        fp.write("%s\n" % item)
```

### Gráfico realizado en tenforflow.org y traído al Colab para graficarlo

```python
from sklearn.manifold import TSNE
import plotly.express as px

# Cargar vectores (sin encabezado):
vectors = pd.read_csv("vectors.tsv", sep="\t", header=None)

# Cargar etiquetas
metadata = pd.read_csv("metadata-edited.tsv", sep="\t")

# Reducir a 3 dimensiones con t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components=3, random_state=42)
coords = tsne.fit_transform(vectors.values)

# Crear DataFrame para graficar
df_plot = pd.DataFrame(coords, columns=["x", "y", "z"])
df_plot["label"] = metadata.iloc[:, 0]  # usa la primera columna como etiqueta

# Graficar en 3D
fig = px.scatter_3d(df_plot, x="x", y="y", z="z", color="label",
                    title="Visualización 3D de Embeddings",
                    width=800, height=600)
fig.show()
```
